```python
#!pip install pypandoc

import shutil
from google.colab import drive
drive.mount('/content/gdrive')

import os
import tempfile
from google.colab import files
import pypandoc
from PyPDF2 import PdfReader, PdfWriter
from reportlab.pdfgen import canvas

def create_unique_watermark(input_pdf, output_pdf, watermark_text):
    with tempfile.NamedTemporaryFile(suffix=".pdf") as watermark_file:
        c = canvas.Canvas(watermark_file.name)
        c.setFont("Helvetica", 12)
        c.drawString(15, 15, watermark_text)
        c.save()

        watermark = PdfReader(watermark_file.name)
        pdf = PdfReader(input_pdf)
        pdf_writer = PdfWriter()

        for page in range(len(pdf.pages)):
            pdf_page = pdf.pages[page]
            pdf_page.merge_page(watermark.pages[0])
            pdf_writer.add_page(pdf_page)

        with open(output_pdf, "wb") as output_file:
            pdf_writer.write(output_file)

def convert_doc_to_pdf(doc_path, pdf_path):
    output = pypandoc.convert_file(doc_path, 'pdf', outputfile=pdf_path, extra_args=['--pdf-engine=xelatex'])
    assert output == ""

def add_watermark_and_save(input_pdf, output_pdf, watermark_text):
    create_unique_watermark(input_pdf, output_pdf, watermark_text)

def get_output_filename():
    output_filename = input("Enter the name for the output PDF file (without .pdf extension): ")
    return f"{output_filename}.pdf"

print("Upload your .docx file:")
uploaded = files.upload()

if not uploaded:
    print("No file was uploaded. Please try again.")
else:
    print("Uploaded files:", uploaded.keys())
    for doc_file in uploaded.keys():
        temp_pdf_file = "temp_output.pdf"
        final_pdf_file = get_output_filename()
        unique_tag = "ID: Massimiliano Pablo Pianges"

        # Create a temporary sanitized copy of the uploaded file
        sanitized_filename = "sanitized_doc_file.docx"
        shutil.copy(doc_file, sanitized_filename)

        print("Converting", sanitized_filename)
        convert_doc_to_pdf(sanitized_filename, temp_pdf_file)
        print("Adding watermark")
        add_watermark_and_save(temp_pdf_file, final_pdf_file, unique_tag)

        # Clean up temporary files
        os.remove(sanitized_filename)
        os.remove(temp_pdf_file)

def save_to_drive_and_download(output_pdf):
    print("Download your converted PDF file:")
    files.download(output_pdf)

    drive_folder = "/content/gdrive/My Drive/Pdf Converted/"
    os.makedirs(drive_folder, exist_ok=True)
    destination = os.path.join(drive_folder, output_pdf)
    shutil.move(output_pdf, destination)
    print(f"The file '{output_pdf}' has been saved to your Google Drive.")


    save_to_drive_and_download(final_pdf_file)
    print("Download your converted PDF file:")

   
    #files.download(final_pdf_file)

    os.remove(final_pdf_file)  # Delete the final PDF file after download
```
